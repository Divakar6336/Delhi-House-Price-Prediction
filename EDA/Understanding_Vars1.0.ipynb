```python
import requests
import bs4
```

```python
content = requests.get('https://en.wikipedia.org/wiki/Neighbourhoods_of_Delhi').text
bs = bs4.BeautifulSoup(content, 'html.parser')
```

```python
sub_divs = []
for idx, e in enumerate(bs.find_all('li')[12:208]):
    sub_divs.append(e.get_text())
```

```python
def set_subdiv(x):
    for sd in sub_divs:
        if 'Rohini' in x.Locality:
            return 'Rohini'
        if 'Dwarka' in x.Locality:
            return 'Dwarka'
        if 'Sultanpur' in x.Locality:
            return 'Sultanpur'
        if 'Budh Vihar'in x.Locality:
            return 'Budh Vihar'
        if 'Mahavir Enclave' in x.Locality:
            return 'Mahavir Enclave'
        if sd in x.Locality:
            return sd
    return 'Other'

df['new_loc'] = df.apply(set_subdiv,axis=1)
```

```python
df.new_loc.value_counts()
```
