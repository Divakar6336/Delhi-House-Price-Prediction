#### List from Understanding_Vars.ipynb

```python
def set_subdiv(x):
    for sd in loc_index:
        if sd in x.Locality:
            return sd
    return 'Other'

df['loc_updated'] = df.apply(set_subdiv,axis=1)
```

```python
g = sns.boxplot(y = 'Price', x = 'loc_updated', data = df)
g = sns.swarmplot(y = 'Price',x = 'loc_updated', data = df, size = 7,color = 'black')
sns.despine()
g.figure.set_size_inches(20,10)
g.set_xticklabels(g.get_xticklabels(), rotation=-65)
plt.show()
```
